```python
import tushare as ts
import pandas as pd
pro = ts.pro_api()
df=ts.get_stock_basics()
```

```python
data = pro.query('stock_basic', exchange='', list_status='L', fields='ts_code,symbol,name,area,industry,list_date')
df2 = pd.DataFrame()
for code in data['ts_code']:
    df = pro.daily(ts_code=code,start_date='20200915', end_date='20200929')
    df1 = pd.DataFrame(data = df, columns = ['ts_code','trade_date','change'])
    for change in df['change']:
        if change >= 10:
            df2 = df2.append(df1)
            print(df2)
   
print('end..............................')
df2.to_csv('222222.csv')
    
```
